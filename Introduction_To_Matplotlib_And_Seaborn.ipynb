```python
import matplotlib.pyplot as plt  #Sublibrary loading
```

```python
#Data
x=[1,2,3,4,5]
y=[2,3,5,7,11]

#marker=o,*,x,P,s
plt.plot (x,y,ls=":",color="#6D2337",lw="3",marker="P",mfc="Orange",mec="r",ms=4) #dotted(:),dashed(--),solid(-),dashed-dotted(-.),None(" ")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.show()
```

```python
import numpy as np
y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])
plt.plot(y1,color="#E7E5E5")
plt.plot(y2,color="b")
plt.show() #bcoz we only want 1 graph
```

```python
#Scatter Plots
x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.scatter(x,y,color="r",marker="P")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()

```

```python
#Bar Plots
cat=["A","B","C","D"]
val=[4,7,1,8]
plt.bar(cat,val,color="black",width=0.3)  #barh for horizontal and bar for vertical, height instead of width for barh
plt.ylabel('Categories')
plt.xlabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
#histogram for frequency distribution
plt.hist(x,y,color="#6D2337",ls=":",lw=0.3)
plt.hist(x,bins=4)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Histogram')
plt.show()
```

Creating Subplots with MAtplot

```python
x=[1,2,3,4,5]
y1=[2,3,5,7,11]
y2=[1,4,6,8,10]

#Creating Subplots
fig, axs = plt.subplots(2)
axs[0].plot(x,y1)
axs[0].set_title("First Plot")
axs[1].plot(x,y2,"tab:orange")
axs[1].set_title("Second Plot")

#displaying the plot
plt.tight_layout()
plt.show()
```

Adding Annotations

```python
x=[1,2,3,4,5]
y1=[2,3,5,7,11]

#Creating a plot with annotations
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plot with Annotations")
plt.annotate("Peak",xy=(5,11),xytext=(3.5,10),arrowprops=dict(facecolor="black",shrink=0.1) )

plt.show()
```

Creating a Line Plot with Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Seaborn build on top of matplotlib, used to beautify

x=[1,2,3,4,5]
y=[2,3,5,7,11]

sns.lineplot(x=x,y=y)
plt.title("Line Plot with SNS")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
sns.scatterplot(x=x,y=y,color="red")
plt.title("Scatter Plot with SNS")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
sns.barplot(x=cat,y=val,width=0.1,color="black")
plt.title("Bar Plot with SNS")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
data=[1,2,2,3,4,3,6,7,1,10]
sns.histplot(data,bins=4)
plt.title("Histogram with SNS")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris_data = sns.load_dataset("iris")
# iris_data.head()

#Creaing a pair plot
sns.pairplot(iris_data,hue="species")
plt.title("Pair Plot with Seaborn")
plt.show()
```
